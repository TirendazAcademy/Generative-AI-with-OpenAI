```python
from io import BytesIO

byte_stream = BytesIO()
```

```python
image.save(byte_stream, format="PNG")
byte_array = byte_stream.getvalue()
```

```python
response = client.images.create_variation(
    image = byte_array,
    n = 2,
    size = "1024x1024",
)
```
